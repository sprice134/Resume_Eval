```python
import ollama

def createQuestion(question, reverseOrder=False):
    resumes = ['Aria Fischer - Junior Software Engineer.txt',
            'Elara Donovan - Junior Software Engineer.txt',
            'Niko Harper - Junior Software Engineer.txt',
            'Rowan Blake - Junior Software Engineer.txt',
            'Siena Randolph - Junior Software Engineer.txt']
    if reverseOrder:
        resumes.reverse()
    prompt = '[Context]: Here are the resumes of 5 candidates. Based on their experience, answer a question at the end:\n'
    
    for i in resumes:
        with open(f'textResumes/{i}', 'r') as file:
            # Read the entire file into a single string
            content = file.read()
        # print(content)
        prompt += content + ('-' * 30) + '\n'
    
    prompt += f'\n[Instructions]: {question}'
    return prompt
```

```python

response = ollama.chat(model='llama3.1', messages=[
    {
        'role': 'user',
        'content': createQuestion('How many candidates did you see?'),
        
    }],
    options={'num_ctx': 4096})

print(response)
print(len(createQuestion('How many applicants were you see and what where there names?')))
print(createQuestion('How many applicants were you see and what where there names?'))
```
